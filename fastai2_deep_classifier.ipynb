```python
import requests
def search_ddg(query):
    url = "https://api.duckduckgo.com/"
    params = {"q": query, "format": "json"}
    response = requests.get(url, params=params)
    return response.json()
results = search_ddg("Python programming")
print(results["RelatedTopics"])
```

```python
def image_search(query):
    url = "https://api.duckduckgo.com/"
    params = {"q": query, "format": "json", "ia": "images"}
    response = requests.get(url, params=params)
    return response.json()
```

```python
image_search("orange")
```

```python
image_results = image_search("orange")
print(image_results["Image"])
```

```python
import requests
import os.path

save_path = 'C:/Users/ninad/Documents/Data-code/Projects/fastai2_deep_classifier/bear_images/'


def download_images(url_list):
    try:
        for image in url_list:
            filename = image.split('/')[-1]
            completeName = os.path.join(save_path, filename)
            r = requests.get(image, allow_redirects=True)
            open(completeName, 'wb').write(r.content)
    except:
        pass # doing nothing on exception
```

```python
download_images(url_list)
```
